```python
import matplotlib.pyplot as plt

# Reusing your provided data
models = ['LSTM', 'Bi-LSTM', 'Transformer', 'ConvTransformer']

# MPJPE scores
mpjpe_scores = {
    'baseline': [0.1130, 0.23309, 0.21089, 0.1063],
    'bio_constraint': [0.1159, 0.18972, 2.01275, 0.1075],
   # 'bio_constraint_pre': [0.33309, 0.33309, 0.31089, 0.3418]
}

# MPVE scores
mpve_scores = {
    'baseline': [0.0048, 0.00423, 0.0099, 0.0024],
    'bio_constraint': [0.0048, 0.01258, 0.14092, 0.0024],
   # 'bio_constraint_pre': [0.01423, 0.01423, 0.0199, 0.1604]
}

# Helper function to add text labels to points
def add_labels(x, y, offset=0.002):
    for i, val in enumerate(y):
        plt.text(i, val + offset, f'{val:.3f}', ha='center', fontsize=8)

# Plot MPJPE
plt.figure(figsize=(7, 5))
for label, color in zip(mpjpe_scores, ['grey', 'blue', 'red']):
    plt.plot(models, mpjpe_scores[label], marker='o', color=color, label=label.replace('_', ' ').title())
    add_labels(models, mpjpe_scores[label], offset=0.01)
plt.title('MPJPE Comparison Across Models')
plt.xlabel('Model')
plt.ylabel('MPJPE (Lower is better)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot MPVE
plt.figure(figsize=(7, 5))
for label, color in zip(mpve_scores, ['grey', 'blue', 'red']):
    plt.plot(models, mpve_scores[label], marker='o', color=color, label=label.replace('_', ' ').title())
    add_labels(models, mpve_scores[label], offset=0.003)
plt.title('MPVE Comparison Across Models')
plt.xlabel('Model')
plt.ylabel('MPVE (Lower is better)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
